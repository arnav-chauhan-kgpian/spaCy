```python
import spacy
from spacy.matcher import Matcher
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
matcher = Matcher(nlp.vocab)
pattern = [{"LIKE_EMAIL":True}]
matcher.add("EMAIL_ADDRESS",[pattern])
```

```python
doc = nlp("My email address in tarzan_car@gmail.com")
matches = matcher(doc)
```

```python
for token in doc:
  print(token)
```

```python
print(matches) #output: [lexeme,start_token,end_token]
print(nlp.vocab[matches[0][0]].text)
```

```python
with open("/content/wiki_mlk.txt","r") as f:
  text = f.read()
```

```python
print(text)
```

```python
#print all proper nouns in the text
nlp = spacy.load("en_core_web_sm")
matcher = Matcher(nlp.vocab)
pattern = [{"POS":"PROPN","OP":"+"}]
matcher.add("PROPER_NOUNS",[pattern],greedy="LONGEST")
doc = nlp(text)
matches = matcher(doc)
matches.sort(key = lambda x:x[1])
print(matches)
```

```python
print(len(matches))
```

```python
for match in matches[:20]:
  print(match,doc[match[1]:match[2]])
```

```python
#suppose we are asked to generate proper nouns followed by a verb
#print all proper nouns in the text
nlp = spacy.load("en_core_web_sm")
matcher = Matcher(nlp.vocab)
pattern = [{"POS":"PROPN","OP":"+"},{"POS":"VERB"}]
matcher.add("PROPER_NOUNS",[pattern],greedy="LONGEST")
doc = nlp(text)
matches = matcher(doc)
matches.sort(key = lambda x:x[1])
print(len(matches))
for match in matches[:20]:
  print(match,doc[match[1]:match[2]])
```

```python
import json
with open("/content/alice.json","r") as f:
  data = json.load(f)
```

```python
text = data[0][2][0]
print(text)
```

```python
text = text.replace("`","'")
print(text)
```

```python
matcher = Matcher(nlp.vocab)
speak_lemmas = ["think","say"]
pattern = [{"ORTH":"'"},
           {"IS_ALPHA":True,"OP":"+"},
           {"IS_PUNCT":True,"OP":"*"},
           {"ORTH":"'"},
           {"POS":"VERB","LEMMA":{"IN":speak_lemmas}},
           {"POS":"PROPN","OP":"+"},
           {"ORTH":"'"},
           {"IS_ALPHA":True,"OP":"+"},
           {"IS_PUNCT":True,"OP":"*"},
           {"ORTH":"'"}]
matcher.add("PROPER_NOUNS",[pattern],greedy="LONGEST")
doc = nlp(text)
matches = matcher(doc)
matches.sort(key = lambda x:x[1])
print(len(matches))
for match in matches[:20]:
  print(match,doc[match[1]:match[2]])
```

```python
matcher = Matcher(nlp.vocab)
speak_lemmas = ["think","say"]
pattern1 = [{"ORTH":"'"},{"IS_ALPHA":True,"OP":"+"},{"IS_PUNCT":True,"OP":"*"},{"ORTH":"'"},{"POS":"VERB","LEMMA":{"IN":speak_lemmas}},{"POS":"PROPN","OP":"+"},{"ORTH":"'"},{"IS_ALPHA":True,"OP":"+"},{"IS_PUNCT":True,"OP":"*"},{"ORTH":"'"}]
pattern2 = [{'ORTH': "'"}, {'IS_ALPHA': True, "OP": "+"}, {'IS_PUNCT': True, "OP": "*"}, {'ORTH': "'"}, {"POS": "VERB", "LEMMA": {"IN": speak_lemmas}}, {"POS": "PROPN", "OP": "+"}]
pattern3 = [{"POS": "PROPN", "OP": "+"},{"POS": "VERB", "LEMMA": {"IN": speak_lemmas}}, {'ORTH': "'"}, {'IS_ALPHA': True, "OP": "+"}, {'IS_PUNCT': True, "OP": "*"}, {'ORTH': "'"}]
matcher.add("PROPER_NOUNS",[pattern1,pattern2,pattern3],greedy="LONGEST")
for text in data[0][2]:
  text = text.replace("`","'")
  doc = nlp(text)
  matches = matcher(doc)
  matches.sort(key = lambda x:x[1])
  print(len(matches))
  for match in matches[:10]:
    print(match,doc[match[1]:match[2]])
```
